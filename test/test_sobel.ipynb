```python
# This is a test for different computations of the Sobel filter

import numpy as np
from scipy import signal, ndimage
from skimage import color, io
from matplotlib import pyplot as plt
```

```python
img = color.rgb2gray(io.imread('bbb.png'))
plt.imshow(img, cmap='gray')
print(img.shape)
```

```python
sob_x = ndimage.sobel(img, axis=0)
sob_y = ndimage.sobel(img, axis=1)
gradient = np.hypot(sob_x, sob_y)
plt.imshow(gradient, cmap="gray")
si = gradient.std()
print(gradient.max())
print(gradient.shape)
print(si)
```

```python
g = np.asarray([[-1, 0, 1],
                [-2, 0, 2],
                [-1, 0, 1]])
sob_x = signal.convolve2d(img, g, mode="valid")
sob_y = signal.convolve2d(img, np.flip(g.T, axis=0), mode="valid")
gradient = np.hypot(sob_x, sob_y)
plt.imshow(gradient, cmap="gray")
si = gradient.std()
print(gradient.max())
print(gradient.shape)
print(si)
```
